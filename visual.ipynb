```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv("dataset (1).csv")
df
```

## histogram

```python
plt.figure(figsize=(10,6))
sns.histplot(df["age"],kde=True,bins=30,color='purple', edgecolor='black',alpha=0.9)
plt.title("distribution of age")
plt.xlabel("age")
plt.ylabel("frequency")
plt.grid(True)
plt.xlim(0,100)
plt.ylim(0,40)
plt.show()
```

## scatter plot

```python
plt.figure(figsize=(5,5))
sns.scatterplot(x='age',y='chol',data=df,color='blue',edgecolor='black',alpha=0.8,s=100,marker=".")
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='trestbps',y='chol',data=df,color='purple',edgecolor='black',alpha=0.8,s=100,marker=".")
plt.grid(True)
plt.show()
```

## line plot

```python
plt.figure(figsize=(6,6))
sns.lineplot(x='age',y='chol',data=df,color='grey', marker="." , markersize= 8 ,alpha=0.8)
plt.gca().set_facecolor("lightblue")
plt.title("age v/s cholestrol")
plt.xlabel("age")
plt.ylabel("cholestrol")
plt.grid(True)
plt.show()
```

## box plot

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='age',data=df,color='gray')
plt.gca().set_facecolor("lightyellow")
plt.title("box plot of age")
plt.xlabel("age")
plt.ylabel("frequency")
plt.grid(True)
plt.show()
```

## pie chart

```python
plt.figure(figsize=(5,5))
plt.pie(df['target'].value_counts(),
labels=['no Disease','Disease'],
colors=['gray','lightcoral'],
autopct='%1.1f%%',
startangle=90,
shadow=True,
wedgeprops={'edgecolor':'black','linewidth':2},
textprops={'fontsize':12,'color':'black'})
plt.gca().set_facecolor("lightcoral")
plt.title("distribution of target variable")
plt.show()
```

## bar graph

```python
sns.set_style("whitegrid")
plt.figure(figsize=(10,6))

# Scatter plot with regression line
sns.regplot(
    data=df,
    x="age",
    y="chol",
    scatter_kws={'s':70, 'alpha':0.6},
    line_kws={'color':'red', 'linewidth':2}
)

# Add title and labels
plt.title("Age vs Cholesterol Level", fontsize=16, fontweight='bold')
plt.xlabel("Age", fontsize=12)
plt.ylabel("Cholesterol Level", fontsize=12)

# Add subtle background color
plt.gca().set_facecolor("#f8f9fa")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for cholesterol levels
labels = ['Desirable (<200 mg/dL)', 'Borderline High (200-239 mg/dL)', 'High (≥240 mg/dL)']
sizes = [45, 30, 25]  # Example percentages
colors = ['#4CAF50', '#FFC107', '#F44336']  # Green, Amber, Red
explode = (0.05, 0.05, 0.1)  # Slightly "pop out" each slice

# Create pie chart
plt.figure(figsize=(8,8))
plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',
    startangle=140,
    explode=explode,
    shadow=True,
    wedgeprops={'edgecolor':'white','linewidth':2}
)

# Title with style
plt.title("Cholesterol Levels Distribution", fontsize=16, fontweight='bold', color='#333')

# Display chart
plt.show()
```
